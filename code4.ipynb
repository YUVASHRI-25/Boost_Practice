```python
df = pd.read_csv("E:/PYTHON/boost_practice/healthcare-dataset-stroke-data.csv")
df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier

y = df["stroke"]
X = df.drop(["stroke", "id"], axis=1)

```

```python

X["bmi"].fillna(X["bmi"].mean(), inplace=True)

```

```python
#Encode categorical variables 
cat_cols = ["gender", "ever_married", "work_type", "Residence_type", "smoking_status"]
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

```

```python
#  Scale numerical features 
num_cols = ["age", "avg_glucose_level", "bmi"]
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])

```

```python

model = GradientBoostingClassifier(random_state=42)

#  Cross-validation 
acc = cross_val_score(model, X, y, cv=5, scoring="accuracy").mean()
roc_auc = cross_val_score(model, X, y, cv=5, scoring="roc_auc").mean()

print("Accuracy:", acc)
print("ROC-AUC:", roc_auc)
```
